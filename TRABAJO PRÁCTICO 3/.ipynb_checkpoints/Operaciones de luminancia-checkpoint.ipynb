```python
#im1 = imageio.imread('Img_oscura.bmp')
im1 = imageio.imread('fuente.bmp')
plt.imshow(im1)
plt.show()
```

```python
imagenNormalizada=ope.normalize_rgb(im1)
plt.imshow(imagenNormalizada)
plt.show()
```

```python

# Genera un histograma con 10 columnas
histogram, bins = np.histogram(imagenNormalizada.flatten(), bins=10, range=(0, 1))

# Crear un gráfico de barras para el histograma
plt.bar(bins[:-1], histogram, width=0.1, align='center', alpha=0.7)
plt.title("Histograma")
plt.xlabel("Valor de píxel normalizado")
plt.ylabel("Frecuencia")
plt.xticks(bins[:-1])
plt.grid(True)
plt.show()

```

```python
YIQ=ope.RGB_to_YIQ(imagenNormalizada)
# Extraer la componente de luminancia (Y)
Y = YIQ[:, :, 0]
import numpy as np
Yprima = Y
# Aplicar la raíz cuadrada para aclarar la luminancia
#Yprima = np.sqrt(Y)
#Yprima = Y*Y
Ymín=0.2
Ymáx=0.8
Yprima[Y<Ymín]=0
Yprima[Y>Ymáx]=1

# Reemplazar la componente Y con la versión aclarada
YIQ[:, :, 0] = Yprima
result=ope.YIQ_to_RGB(YIQ)
resultmostrar=ope.RGB_to_bytes(result)
plt.imshow(resultmostrar)
plt.show()
```

```python
# Genera un histograma con 10 columnas
histogram, bins = np.histogram(result.flatten(), bins=10, range=(0, 1))
# Crear un gráfico de barras para el histograma
plt.bar(bins[:-1], histogram, width=0.1, align='center', alpha=0.7)
plt.title("Histograma")
plt.xlabel("Valor de píxel normalizado")
plt.ylabel("Frecuencia")
plt.xticks(bins[:-1])
plt.grid(True)
plt.show()
```
